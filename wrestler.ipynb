```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Read the file using the read_csv method for the year from 2013-2016
df13 = pd.read_csv("../../GWDC201805DATA3-Class-Repository-DATA/05-Matplotlib/2/Activities/08-Stu_WinnerWrestling-Part1/Resources/WWE-Data-2013.csv")
df14 = pd.read_csv("../../GWDC201805DATA3-Class-Repository-DATA/05-Matplotlib/2/Activities/08-Stu_WinnerWrestling-Part1/Resources/WWE-Data-2014.csv")
df15 = pd.read_csv("../../GWDC201805DATA3-Class-Repository-DATA/05-Matplotlib/2/Activities/08-Stu_WinnerWrestling-Part1/Resources/WWE-Data-2015.csv")
df16 = pd.read_csv("../../GWDC201805DATA3-Class-Repository-DATA/05-Matplotlib/2/Activities/08-Stu_WinnerWrestling-Part1/Resources/WWE-Data-2016.csv")
```

```python
df14.head(2)
```

```python
# Starting merging win, draw and loss
df23 = pd.merge(df13, df14, on= "Wrestler", how = "outer", suffixes=("_13","_14"))
df24 = pd.merge(df15, df16, on = "Wrestler", how ="outer", suffixes=("_15","_16"))
```

```python
df33 = pd.merge(df23,df24, on ="Wrestler", how ="outer")
df33 = df33.fillna(0)
```

```python
wrest = df33.set_index("Wrestler", drop = True)
```

```python
# Display the combined data of the wrestler
wrest.head(30)
```

```python
# Choose the wrestler name and display the performance in pie-chart

choice = input("Choose a wrestler")
win = wrest.loc[[choice],["Wins_13","Wins_14", "Wins_15","Wins_16"]].sum(axis =1)
loss = wrest.loc[[choice],["Losses_13","Losses_14", "Losses_15","Losses_16"]].sum(axis =1)
draw = wrest.loc[[choice],["Draws_13","Draws_14", "Draws_15","Draws_16"]].sum(axis =1)
total = [win,loss,draw]
color =["green","red","yellow"]
fig = plt.figure(figsize=(15,8))
plt.pie(total, labels=["win","loss","draw"],autopct="%1.1f%%",colors = color, shadow=False, startangle=140)
plt.title(f"Performance of Wrestler- {choice}", fontsize=17, fontweight = "bold")
plt.grid()
plt.axis("equal")
```
